```python
# Import necessary libraries
import numpy as np
import tensorflow as tf
from tensorflow.keras.applications import VGG16
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import img_to_array, load_img
from sklearn.model_selection import train_test_split
import os
import matplotlib.pyplot as plt

# Set inline plotting for Jupyter
%matplotlib inline

# Define paths for data
# Assuming you have directories 'DigitalText' and 'HandWritten' with corresponding images
digital_text_dir = '/Users/rahul/Downloads/SUMMER INTERNSHIP/HWvsDT/training_dataset/DT'  # Replace with your directory path
handwritten_text_dir = '/Users/rahul/Downloads/SUMMER INTERNSHIP/HWvsDT/training_dataset/HW'  # Replace with your directory path

# Load images and labels
def load_images_from_folder(folder, label):
    images = []
    labels = []
    for filename in os.listdir(folder):
        img_path = os.path.join(folder, filename)
        try:
            img = load_img(img_path, target_size=(64, 64))  # VGG16 expects 3-channel RGB images
            img_array = img_to_array(img) / 255.0  # Normalize pixel values
            images.append(img_array)
            labels.append(label)
        except Exception as e:
            print(f"Error loading image {img_path}: {e}")
    return images, labels

digital_images, digital_labels = load_images_from_folder(digital_text_dir, 0)  # Label '0' for DigitalText
handwritten_images, handwritten_labels = load_images_from_folder(handwritten_text_dir, 1)  # Label '1' for HandWritten

# Combine and split data
images = np.array(digital_images + handwritten_images)
labels = np.array(digital_labels + handwritten_labels)

X_train, X_test, y_train, y_test = train_test_split(images, labels, test_size=0.2, random_state=42)

# Define the VGG16-based model for transfer learning
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(64, 64, 3))

# Freeze the layers in the base model
base_model.trainable = False

# Build the full model
model = Sequential([
    base_model,
    GlobalAveragePooling2D(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')  # Sigmoid for binary classification
])
```
